```python
df.drop_duplicates(inplace=True)
```

```python
cols = ['BMI','PhysHlth','MentHlth']
```

```python
df[cols].mean().plot(kind='bar')

plt.title("Average Values Before Normalization")

plt.show()
```

```python
plt.figure(figsize=(7,5))

df[cols].boxplot()

plt.title("Before Normalization")

plt.show()
```

```python
plt.plot(df['BMI'][:100])

plt.title("BMI Before Normalization")

plt.xlabel("Records")

plt.ylabel("BMI")

plt.show()
```

```python
minmax = MinMaxScaler()
```

```python
df_minmax = df.copy()
```

```python
df_minmax[cols] = minmax.fit_transform(df[cols])
```

```python
comparison = pd.DataFrame({

'Original BMI':df['BMI'].head(),

'Normalized BMI':df_minmax['BMI'].head()

})

comparison
```

```python
df_minmax[cols].boxplot()

plt.title("After Min-Max Normalization")

plt.show()
```

```python
std = StandardScaler()
```

```python
df_z = df.copy()
```

```python
df_z[cols] = std.fit_transform(df[cols])
```

```python
df_z[cols].mean()
```

```python
df_z[cols].std()
```

```python
df_z[cols].boxplot()

plt.title("After Z Score Normalization")

plt.show()
```
